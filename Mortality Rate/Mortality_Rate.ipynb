# **MORTALITY RATES IN THE PHILIPPINES**

```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import io
from matplotlib import pyplot as plt
from google.colab import data_table
```

```python
data= pd.read_csv('https://raw.githubusercontent.com/Glennarcega/IT-BA-3203-PROJECT/c7210edc449ad25457dc1954a4d81cea8f5191ca/Mortality%20Rate/Infant.csv')
data.head(300)
```

Here, we are presenting the average mortality rate of infants in Philippines from 2006 to 2015.

```python
Year = data['Year']
Mortality = data['PHILIPPINES']


fig = plt.figure(figsize=(15,6))
plt.plot(Year, Mortality)
plt.xlabel('Years')
plt.ylabel('Infant Mortality Rate (%) ')
plt.title('Average Mortality Rate Of Infants of Different Regions In Philippines 2006-2015')
plt.show()
```

```python
categories = ['Batangas', 'Cavite', 'Laguna', 'Rizal', 'Quezon']
rate = [14.1, 14.8, 13.4, 14.4, 13.6]

plt.pie(rate, labels=categories, autopct='%.1f%%', colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#CF9FFF'])
plt.axis('equal')
plt.title('Mortality Rate (%) of Infants CALABARZON in 2015')
```
